```python
import pandas as pd
import numpy as np 
```

```python
file1 = pd.read_csv('file1.csv')
file1.head()
file2 = pd.read_csv('file2.csv')
file3 = pd.read_csv('file3.csv')
```

```python
print(file1.shape)
print(file2.shape)
print(file3.shape)
```

```python
print(file1.columns)
print(file2.columns)
print(file3.columns)
```

```python
file3 = file3.rename(columns={'State':'ST', 'Gender':'GENDER'})
file3.columns
```

```python
file2 = file2.reindex(columns=['Customer', 'ST', 'GENDER', 'Education', 'Customer Lifetime Value',
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Policy Type', 'Vehicle Class', 'Total Claim Amount'])

file3 = file3.reindex(columns=['Customer', 'ST', 'GENDER', 'Education', 'Customer Lifetime Value',
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Policy Type', 'Vehicle Class', 'Total Claim Amount'])
print(file2.columns)
print(file3.columns)
```

```python
data = pd.concat([file1, file2, file3], axis=0)
data.shape
```

```python
data.select_dtypes(include='number')
#These are numerical columns
```

```python
data.select_dtypes(include='object')
#These are categorical columns
```

```python
data = data.drop(['Education', 'Number of Open Complaints'], axis=1)
data.head()
```

```python
data['Customer Lifetime Value'] = data['Customer Lifetime Value'].str.replace('%', '')
data
```

```python
data['Customer Lifetime Value'] = pd.to_numeric(data['Customer Lifetime Value'])
data['Customer Lifetime Value'] = list(map(lambda x: x / 100, data['Customer Lifetime Value']))
data
```

```python
data.duplicated()
print(data)
```

```python
data = data.drop_duplicates()
print(data)
```
